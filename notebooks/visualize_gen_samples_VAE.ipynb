```python
import sys
import os
sys.path.append(os.path.dirname(os.getcwd()))
from src.models.vanilla_vae import ConvVAE
import torch
import matplotlib.pyplot as plt
```

```python
model = ConvVAE()
model = torch.load('/home/syurtseven/gsoc/reports/exp/2/model/model_2.pt')
```

```python


# Define the sampling function
def sample_from_vae(model, num_samples=1, latent_dim=20):
    with torch.no_grad():
        z = torch.randn(num_samples, latent_dim).to('cuda:0')
        generated_samples = model.decode(z)
        
        return generated_samples

# Generate samples
num_samples = 25  # Define the number of samples you want to generate
samples1 = sample_from_vae(model, num_samples=num_samples)
samples2 = sample_from_vae(model, num_samples=num_samples)
```

```python
# Visualize the samples
# def visualize_samples(samples):
#     samples = samples.cpu().numpy()
#     fig, axes = plt.subplots(1, num_samples, figsize=(15, 3))
#     for i, ax in enumerate(axes):
#         ax.imshow(samples[i, 0])  # Assuming single-channel images
#         ax.axis('off')
#     plt.show()
def visualize_samples(samples):
    samples = samples.cpu().numpy()
    fig, axes = plt.subplots(5, 5, figsize=(15, 15))  # Create a 5x5 grid of subplots
    for i in range(5):
        for j in range(5):
            index = i * 5 + j
            if index < len(samples):
                axes[i, j].imshow(samples[index, 0])  # Assuming single-channel images
                axes[i, j].axis('off')
            else:
                axes[i, j].axis('off')  # Turn off axes for any remaining empty plots
    plt.tight_layout()
    plt.show()
# Call the visualization function
visualize_samples(samples1)
```
