```python
cases = glob.glob('/home/syurtseven/GI-Tract-Image-Segmentation/synthetic_data_generation/gsoc/data/UWMadison_kaggle/train/*')
```

```python
num_days   = []
num_slices = []
that = []
count =0
for case in cases:
    days = glob.glob(os.path.join(case,'*'))
    #print(f'{case.split('/')[-1]} has {len(days)} days')
    num_days.append(len(days))
    for day in days:
        slices = glob.glob(os.path.join(day,'*','*'))
        #print(f'{day.split('_')[-1]} has {len(slices)} slices')
        num_slices.append(len(slices))
        for slice in slices:
            re = cv2.imread(slice, cv2.IMREAD_UNCHANGED)
            if len(re.shape) != 2 :
                that.append(slice)
                count +=1

count    
```
